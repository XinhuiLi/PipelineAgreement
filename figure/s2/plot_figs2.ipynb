```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

LOCAL_ROOT = os.environ.get('LOCAL_ROOT')
```

```python
type='power'
hbn_afni_good=np.load(f'{LOCAL_ROOT}/PipelineHarmonization/figure/s2/HBN_AFNI_FD_'+type+'_good.npy')
hbn_afni_bad=np.load(f'{LOCAL_ROOT}/PipelineHarmonization/figure/s2/HBN_AFNI_FD_'+type+'_bad.npy')
hbn_fsl_good=np.load(f'{LOCAL_ROOT}/PipelineHarmonization/figure/s2/HBN_FSL_FD_'+type+'_good.npy')
hbn_fsl_bad=np.load(f'{LOCAL_ROOT}/PipelineHarmonization/figure/s2/HBN_FSL_FD_'+type+'_bad.npy')
```

```python
hbn_afni_good_fd_list, hbn_afni_bad_fd_list = [], []
hbn_fsl_good_fd_list, hbn_fsl_bad_fd_list = [], []

for i in range(6):
    hbn_afni_good_fd_list.append(hbn_afni_good[i,:])
    hbn_afni_bad_fd_list.append(hbn_afni_bad[i,:])
    hbn_fsl_good_fd_list.append(hbn_fsl_good[i,:])
    hbn_fsl_bad_fd_list.append(hbn_fsl_bad[i,:])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

y_label = 'FD Power Correlation'
tick_label = ['mean vs median', 'mean vs first', 'mean vs last', 'median vs first', 'median vs last', 'first vs last']

sns.boxplot(data=hbn_afni_good_fd_list, ax=axs[0,0])
axs[0,0].set_ylim([0.58,1.02])
axs[0,0].set_xticklabels(tick_label, rotation=15)
axs[0,0].set_ylabel(y_label)
axs[0,0].set_title('AFNI 3dvolreg low motion', fontsize=15, fontweight='bold')

sns.boxplot(data=hbn_afni_bad_fd_list, ax=axs[0,1])
axs[0,1].set_ylim([0.58,1.02])
axs[0,1].set_xticklabels(tick_label, rotation=15)
axs[0,1].set_ylabel(y_label)
axs[0,1].set_title('AFNI 3dvolreg high motion', fontsize=15, fontweight='bold')

sns.boxplot(data=hbn_fsl_good_fd_list, ax=axs[1,0])
axs[1,0].set_ylim([0.58,1.02])
axs[1,0].set_xticklabels(tick_label, rotation=15)
axs[1,0].set_ylabel(y_label)
axs[1,0].set_title('FSL MCFLIRT low motion', fontsize=15, fontweight='bold')

sns.boxplot(data=hbn_fsl_bad_fd_list, ax=axs[1,1])
axs[1,1].set_ylim([0.58,1.02])
axs[1,1].set_xticklabels(tick_label, rotation=15)
axs[1,1].set_ylabel(y_label)
axs[1,1].set_title('FSL MCFLIRT high motion', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.savefig('FigureS2.1.png', dpi=300)
```

```python
fig, axs = plt.subplots(4, 4, figsize=(15, 12))

for i in range(1,4):
    for j in range(4-i,4):
        axs[i, j].axis('off')

path = f'{os.environ.get("BASEFOLDER")}/ROI/ROI_Schaefer200'
reference = ['mean', 'median', 'first', 'last']

for i, r1 in enumerate(reference):
    for j, r2 in enumerate(reference[i:]):
        p1 = os.path.join(path, 'motion_afni_'+r1)
        p2 = os.path.join(path, 'motion_fsl_'+r2)
        
        p1_list = os.listdir(p1)
        p2_list = os.listdir(p2)

        corr = np.zeros( len(p1_list) )
        for k, sub in enumerate(p1_list):
            roi1 = np.loadtxt( os.path.join(path, 'motion_afni_'+r1, p1_list[k]) )
            roi2 = np.loadtxt( os.path.join(path, 'motion_afni_'+r2, p2_list[k]) )
            c, _ = pearsonr(roi1, roi2)
            corr[k] = c
        
        a = sns.swarmplot(x=corr.T, ax=axs[i,j], size=8, orient='v')
        a.set(xticklabels=[])
        a.set(xlabel=None)
        a.set(ylabel=None)
        # a.set(title='AFNI '+r1+' vs FSL '+r2)
        a.set(ylim=(0.945, 1.005))

fig.tight_layout()
plt.savefig('FigureS2.3.png',bbox_inches='tight',dpi=200)
```
